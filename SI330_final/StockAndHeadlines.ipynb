```python
##Block of code makes a new dataframe that contains the articles
#drop the na rows
#convert dataframe to one large string, make all the words lowercase
#use regex to clean the string to the way i want it

###ISSUE: News Api doesnt let you get full articles, it has some of article then [ ### chars+], so the data doesn't contain full articles

import re
a = full_df[['article', 'article.2', 'article.3', 'article.4', 'article.5', 'article.6', 'article.7', 'article.8', 'article.9']].copy()
art = a.dropna()

article = art.to_string()

v = re.sub(r'\W+',' ', article)
v2 = re.sub(r' article article 2 article 3 article 4 article 5 article 6 article 7 article 8 article 9 214 ', '', v)
v3 = re.sub(r'[A-Z]+ Reuters', '', v2)
v4 = re.sub(r'\d{3,4} chars', '', v3)
v5 = v4.lower()
```

```python
v5_split = v5.split()
```

```python
##Creates a Word2Vec model with the articles
import gensim
articlemodel = gensim.models.Word2Vec([v5_split], min_count = 1)
```

```python
##Saving the values of most similar to a variable to be used later
#added the lists together, to pass into function makeWordCloud
from wordcloud import WordCloud
tri = articlemodel.wv.most_similar('thomson')
reut = articlemodel.wv.most_similar('reuters')

fb = articlemodel.wv.most_similar('facebook')
aapl = articlemodel.wv.most_similar('apple')
amzn = articlemodel.wv.most_similar('amazon')
nflx = articlemodel.wv.most_similar('netflix')

g = articlemodel.wv.most_similar('general')
m = articlemodel.wv.most_similar('motors')

sbux = articlemodel.wv.most_similar('starbucks')
nke = articlemodel.wv.most_similar('nike')
ba = articlemodel.wv.most_similar('boeing')
mcd = articlemodel.wv.most_similar('mcdonalds')

article_similar = tri + reut + fb + aapl + amzn + nflx + g + m + sbux + nke + ba + mcd
```

```python
##Function joins together a list of strings to a single string
def listToString(s):  
    st = " "    
    return (st.join(s))    
## Source: https://www.geeksforgeeks.org/python-program-to-convert-a-list-to-string/

##Function creates the word cloud
#passed in the similar words, append the 1st element to a list
#converts the list of strings to a single string, in order to make the word cloud
#Reference: https://www.datacamp.com/community/tutorials/wordcloud-python
def makeWordCloud(data):
    l = []
    for i in data:
        l.append(i[0])
    l_str = listToString(l)
    
    word = WordCloud(background_color = "white", height = 600, width = 1000, min_font_size = 7).generate(l_str)
    plt.figure()
    plt.imshow(word, interpolation='bilinear')
    plt.axis("off")

makeWordCloud(article_similar)
plt.savefig('wordcloud.png')
```
