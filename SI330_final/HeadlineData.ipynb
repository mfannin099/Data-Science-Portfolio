```python
##Function is passed in the data received from the API
#empty lists are created to save values
#loop through the data that has 'articles' as the key
#append each item i want using its key, and saving the values to its respective list

def extractData(x_articles):
    total = 0
    title = []
    description = []
    date = []
    content = []
    
    total = x_articles['totalResults']
    
    data = x_articles['articles']
    
    for i in data:
        title.append(i['title'])
        description.append(i['description'])
        date.append(i['publishedAt'])
        content.append(i['content'])
    
    
    return total, title, description, date, content
    
    
```

```python
##Calling the funciion to get the data I want

tri_data = extractData(tri_articles) #tri_data[0], tri_data[3] to access the different parts 
fb_data = extractData(fb_articles)
aapl_data = extractData(aapl_articles)
amzn_data = extractData(amzn_articles)
nflx_data = extractData(nflx_articles)
gm_data = extractData(gm_articles)
sbux_data = extractData(sbux_articles)
nke_data = extractData(nke_articles)
ba_data = extractData(ba_articles)
mcd_data = extractData(mcd_articles)
```

```python
##Function is passed in the data that was kept, but only the date 
#it cleans the data, and only keeps the first 10 characters to keep the format: xxxx-xx-xx
#returns the 

def cleanDate (data):
    date = []
    for i in data:
        date.append(i[:10])
    return date
```

```python
## saving new dates

tri_date = cleanDate(tri_data[3])
fb_date = cleanDate(fb_data[3])
aapl_date = cleanDate(aapl_data[3])
amzn_date = cleanDate(amzn_data[3])
nflx_date = cleanDate(nflx_data[3])
gm_date = cleanDate(gm_data[3])
sbux_date = cleanDate(sbux_data[3])
nke_date = cleanDate(nke_data[3])
ba_date = cleanDate(ba_data[3])
mcd_date = cleanDate(mcd_data[3])
```

```python
##Creating all the dataframes
#passed in the data selected, the clean date, and a string that serves as the label
#returns a dataframe

def createData (data, data2, company):
    df = pd.DataFrame({'total_entries': data[0], 'article_title': data[1], 'article_description': data[2], 'article': data[4], 'date': data2})
    df[company] = ""
    return df

tri_df = createData(tri_data, tri_date, 'Thomson Reuters')
fb_df = createData(fb_data, fb_date, 'Facebook')
aapl_df = createData(aapl_data, aapl_date, 'Apple')
amzn_df = createData(amzn_data, amzn_date, 'Amazon')
nflx_df = createData(nflx_data, nflx_date, 'Netflix')
gm_df = createData(gm_data, gm_date, 'General Motors')
sbux_df = createData(sbux_data, sbux_date, 'Starbucks')
nke_df = createData(nke_data, nke_date, 'Nike')
ba_df = createData(ba_data, ba_date, 'Boeing')
mcd_df = createData(mcd_data, mcd_date, 'McDonalds')
```

```python
article_d = pd.concat([tri_df, fb_df, aapl_df, amzn_df, nflx_df, gm_df, sbux_df, nke_df, ba_df, mcd_df], axis = 1)
article_d.head()
```
