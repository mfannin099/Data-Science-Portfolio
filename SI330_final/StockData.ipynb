```python
from alpha_vantage.timeseries import TimeSeries
import pandas as pd
```

```python
#data comes in a nested dictionary format
key = ''
ts = TimeSeries(key)
```

```python
##Making the calls to the Alpha Vantage API
#Standard API call frequency is 5 calls per minute and 500 calls per day

tri = ts.get_monthly(symbol='TRI') #Thomson Reuters
fb = ts.get_monthly(symbol='FB') #Facebook
aapl = ts.get_monthly(symbol='AAPL') #Apple
amzn = ts.get_monthly(symbol='AMZN') #Amazon
nflx = ts.get_monthly(symbol='NFLX') #Netflix
```

```python
gm = ts.get_monthly(symbol='GM') #GM
sbux = ts.get_monthly(symbol='SBUX') #Starbucks
nke = ts.get_monthly(symbol='NKE') #Nike
ba = ts.get_monthly(symbol='BA') #Boeing
mcd = ts.get_monthly(symbol='MCD') #McDonald's
```

```python
##Function takes in the data from the API 
#Gets the first dictionary from the data
#creates the dataframe by saving the values and keys of the data to make columns and rows
#returns the resulting dataframe

def makeData (stock):
    data = stock[0]
    stock_df = pd.DataFrame(data.values(), index = data.keys())
    return stock_df   
```

```python
#Creates the DataFrames
tri_df = makeData(tri)
fb_df = makeData(fb)
aapl_df = makeData(aapl)
amzn_df = makeData(amzn)
nflx_df = makeData(nflx)
gm_df = makeData(gm)
sbux_df = makeData(sbux)
nke_df = makeData(nke)
ba_df = makeData(ba)
mcd_df = makeData(mcd)
```

```python
#Adds Company name to the DataFrame to serve as a label
tri_df["Thomson Reuters"] = ""
fb_df["Facebook"] = ""
aapl_df['Apple'] = ""
amzn_df['Amazon'] = ""
nflx_df['Netflix'] = ""
gm_df['GM'] =""
sbux_df['Starbucks'] = ""
nke_df['Nike'] = ""
ba_df['Boeing'] = ""
mcd_df['McDonalds'] = ""
```

```python
##creates the dataframe with all 10 comapanies stock data
#Added the companies name to the right, to identify the comapnies easier
#changed the N/A values to 0's


financial_d = pd.concat([tri_df, fb_df, aapl_df, amzn_df, nflx_df, gm_df, sbux_df, nke_df, ba_df, mcd_df], axis = 1)
financial_data = financial_d.fillna(0).reset_index().rename(columns= {'index':'date'})
financial_data.head()
```
