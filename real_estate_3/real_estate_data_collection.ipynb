```python
import pandas as pd
import json
```

```python
# function that makes the json readable, and extracts the values and returns them

def extractData(city):
    city_json = city.json()
    
    active_neighborhoods = city_json['content']['active_neighborhoods']
    investment_properties = city_json['content']['investment_properties']
    airbnb_listings = city_json['content']['airbnb_listings']
    traditional_listings = city_json['content']['traditional_listings']
    avg_property_price = city_json['content']['avg_property_price']
    avg_price_per_sqft = city_json['content']['avg_price_per_sqft']
    avg_days_on_market = city_json['content']['avg_days_on_market']
    avg_occupancy = city_json['content']['avg_occupancy']
    avg_nightly_price = city_json['content']['avg_nightly_price']
    avg_airbnb_ROI = city_json['content']['avg_airbnb_ROI']
    avg_airbnb_rental = city_json['content']['avg_airbnb_rental']
    avg_traditional_ROI = city_json['content']['avg_traditional_ROI']
    avg_traditional_rental = city_json['content']['avg_traditional_rental']
    
    return active_neighborhoods,investment_properties, airbnb_listings, traditional_listings, avg_property_price, avg_price_per_sqft, avg_days_on_market, avg_occupancy, avg_nightly_price, avg_airbnb_ROI, avg_airbnb_rental, avg_traditional_ROI, avg_traditional_rental     
    
    
```

```python
#calling the function and saving the values, to be used to make a data frame later

york = extractData(new_york)
la = extractData(los_angeles)
houst = extractData(houston)
phil = extractData(philly)
phoe = extractData(phoenix)
san = extractData(san_antonio)
sanD = extractData(san_diego)
dal = extractData(dallas)
sanJ = extractData(san_jose)
ind = extractData(indy)
jack = extractData(jacksonville)
sanF = extractData(san_fran)
aust = extractData(austin)
col = extractData(columbus)
fort = extractData(fort_worth)
lou = extractData(louisville)
char = extractData(charlotte)
det = extractData(detroit)
el = extractData(el_paso)
mem = extractData(memphis)
nash = extractData(nashville)
balt = extractData(baltimore)
bost = extractData(boston)
sea = extractData(seattle)
wash = extractData(washington)
den = extractData(denver)
mil = extractData(milwaukee)
#por = extractData(portland)
vegas = extractData(las_vegas)
```

```python
#function made to make a data frame for each city. Passed in the city data and the name of the city

def createDf (data, city):
    df = pd.DataFrame({'city': city, 'neighborhoods': data[0], 'investment_properties': data[1], 'airbnb_listings': data[2], 'traditional_listings': data[3], 
                       'avg_property_price': data[4],'avg_price_per_sqft': data[5], 'avg_days_on_market': data[6], 'avg_occupancy': data[7],
                      'avg_nightly_price': data[8], 'avg_airbnb_roi': data[9], 'avg_airbnb_rental': data[10], 'avg_traditional_roi': data[11],
                      'avg_traditional_rental': data[12]}, index = [0])
    return df
```

```python
#creating a data frame for each city, and setting the index

yorkDf = createDf(york, 'New York City').set_index('city')
laDf = createDf(la, 'Los Angeles').set_index('city')
houstDf = createDf(houst, 'Houston').set_index('city')
philDf = createDf(phil, 'Philadelphia').set_index('city')
phoeDf = createDf(phoe, 'Phoenix').set_index('city')
sanDf = createDf(san, 'San Antonio').set_index('city')
sanDDf = createDf(sanD, 'San Diego').set_index('city')
dalDf = createDf(dal, 'Dallas').set_index('city')
sanJDf = createDf(sanJ, 'San Jose').set_index('city')
indDf = createDf(ind, 'Indianapolis').set_index('city')
jackDf = createDf(jack, 'Jacksonville').set_index('city')
sanFDf = createDf(sanF, 'San Francisco').set_index('city')
austDf = createDf(aust, 'Austin').set_index('city')
colDf = createDf(col, 'Columbus').set_index('city')
fortDf = createDf(fort, 'Fort Worth').set_index('city')
louDf = createDf(lou, 'Louisville').set_index('city')
charDf = createDf(char, 'Charlotte').set_index('city')
detDf = createDf(det, 'Detroit').set_index('city')
elDf = createDf(el, 'El Paso').set_index('city')
memDf = createDf(mem, 'Memphis').set_index('city')
nashDf = createDf(nash, 'Nashville').set_index('city')
baltDf = createDf(balt, 'Baltimore').set_index('city')
bostDf = createDf(bost, 'Boston').set_index('city')
seaDf = createDf(sea, 'Seattle').set_index('city')
washDf = createDf(wash, 'Washington D.C').set_index('city')
denDf = createDf(den, 'Denver').set_index('city')
milDf = createDf(mil, 'Milwaukee').set_index('city')
vegasDf = createDf(vegas, 'Las Vegas').set_index('city')
```

```python
#making sure the data is loaded in correctly

cities = pd.concat([yorkDf, laDf, houstDf, philDf, phoeDf, sanDf, sanDDf, dalDf, sanJDf, indDf, jackDf, sanFDf, austDf, colDf, fortDf, louDf, charDf, 
                    detDf, elDf, memDf, nashDf, baltDf, bostDf, seaDf, washDf, denDf, milDf, vegasDf], axis = 0)
cities
```

```python
#exporting to csv, in order to load into R

cities.to_csv('real_estate_data.csv')
```
